```python
%matplotlib inline  
#Using matplotlib backend

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x)+2, ':')
plt.plot(x, np.cos(x)+2, '--');

fig.savefig('tmp_files/my_figure.png')

plt.show()
```

```python
plt.subplot(2, 1, 1)     # (rows, columns, panel number)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```
