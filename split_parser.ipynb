```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

class SplitPageParser:
    SEARCH_GROUP_COUNT = 20
    
    def __init__(this,url):
        this.url = url
        try:
            html = urlopen(this.url)
        except HTTPError as e:
            print("HTTPError")
            return None
        try:
            this.bsObj = BeautifulSoup(html.read())
            
            this.tables = this.getTables()
            this.groups = this.getGroups()
        
        except AttributeError as e:
            print("AttributeError")            
            this.bsObj = None                
    
    def getTitle(this):
        try:
            title = this.bsObj.body.h1
            title = title.get_text() 
        except AttributeError as e:
            title = "SFR"
        return title

    def getTables(this):
        tableList = this.bsObj.findAll("table",{"class":"rezult"})
        return tableList
    
    def getTablesNum(this):
        return len(this.tables)
    
    def getTablesRow(this,ind):        
        rez_table = list()        
        table_row = this.tables[ind].findAll("tr")
                
        #print("Table {0}, rows: {1}".format(ind, len(table_row)))

        for row in table_row:
            tr = row.findAll({"th","td"})        
            rez_row = list()
            for t in tr:
                rez_row.append(t.get_text())
            #print(rez_row)
            rez_table.append(rez_row)       
        
        rez_table[0][0] = this.groups[ind]
        
        return rez_table     
 
    def getGroups(this):
        groups = list()
        gr = this.bsObj.findAll("a")
        for g in gr:
            sg = str(g)
            rez = sg.find("name=")
            if rez>0 and len(sg) < this.SEARCH_GROUP_COUNT:
                il = sg.find('"')+1
                ir = sg.rfind('"')       
                groups.append(sg[il:ir])
        this.groups = groups
        return groups
    
    def getGroupsNum(this):
        return len(this.groups)
    
#SPLITS_URL = "http://orient-simf.ru/doc/doc_2019/splits200419.htm"
#SPLITS_URL = "http://orient-crimea.narod.ru/rez/rez_2019/split_ChEvp-2019_04_20.html"

SPLITS_URL = "http://evp-orient.site/splits200419loc.htm"   # localhost
print(SPLITS_URL)
    
p = SplitPageParser(SPLITS_URL)

print(p.getTitle())
print("Number of tables: ", p.getTablesNum())
print("Number of groups: ", p.getGroupsNum())   

table = p.getTablesRow(0)
print(table)
```
