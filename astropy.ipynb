```python
%matplotlib inline
import numpy as np
from scipy.fft import fft
import matplotlib.pyplot as plt
```

```python
N = 2048
T = 1.0/400

x = np.linspace(0,N*T,N)
y = np.sin(150*2*np.pi*x) + \
    np.sin(100*2*np.pi*x)

yfc = fft.rfft(y)
yf = 2./N*np.abs(yfc[0:N//2])

from scipy.signal import blackman
w = blackman(N)
yfcw = fft.rfft(y*w)
yfw = 2./N*np.abs(yfcw[0:N//2])

xf = np.linspace(0,1.0/(2.*T),N//2)

#plt.plot(xf,yf,xf,yfw)
plt.semilogy(xf[1:],yf[1:],xf[1:],yfw[1:])
plt.legend(['FFT','FFT blackman'])
plt.show()
```

```python
fig,axes = plt.subplots(figsize=[16,8], ncols = 2,nrows = 1)
axes[0].plot(x,y)
axes[0].set_xlim(0,N*T/10)
axes[0].set_title('$\sin (2\pi x)',fontsize=20)
axes[1].semilogy(xf[1:],yf[1:],xf[1:],yfw[1:])
plt.show()
```
